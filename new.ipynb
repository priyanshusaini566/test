# problem-7

```python
n=int(input("Enter the number : "))
fact=1

for i in range(n,0,-1):
    fact=fact*i

print(f"the factorial of {n} will be : ",fact)
```

# question-11

```python
import math
l1=input()
l2=[]
l3=l1.split(',')

for i in range(len(l3)):
    result=math.sqrt((2*int(l3[i])*50)/30)
    l2.append(str(int(result)))

print(','.join(l2))


```

# question-12

```python
input_str = input()
dimensions=[int(x) for x in input_str.split(',')]
rowNum=dimensions[0]
colNum=dimensions[1]
multilist = [[0 for col in range(colNum)] for row in range(rowNum)]

for row in range(rowNum):
    for col in range(colNum):
        multilist[row][col]= row*col

print(multilist)
```
